```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.init_printing()
```

oscylator harmoniczny z tłumieniem

```python
x = sp.Function('x',real=True)
t = sp.Symbol('t',real=True)
w0 = sp.Symbol('omega_0',real=True,positive=True)
z = sp.Symbol('zeta',real=True)

x0 = sp.Symbol('x_0',real=True)
v0 = sp.Symbol('v_0',real=True)


x(t),t,w0,z,x0,v0
```

```python
sol = sp.dsolve(x(t).diff(t,2)+2*z*w0*x(t).diff(t)+w0**2*x(t),0,ics={x(0):x0,x(t).diff(t).subs(t,0):v0})
sol
```

```python
nfun1 = sp.lambdify(t,sol.rhs.subs({x0:10,v0:-1,w0:2,z:0.04}))
nfun12 = sp.lambdify(t,sol.rhs.subs({x0:10,v0:-1,w0:2,z:0.1}))
nfun13 = sp.lambdify(t,sol.rhs.subs({x0:10,v0:-1,w0:2,z:0.5}))
nfun14 = sp.lambdify(t,sol.rhs.subs({x0:10,v0:-1,w0:2,z:1.2}))
```

```python
ts = np.linspace(0,15,1000)
plt.plot(ts,np.real(nfun1(ts)),label='zeta=0.04')
plt.plot(ts,np.real(nfun12(ts)),label='zeta=0.1')
plt.plot(ts,np.real(nfun13(ts)),label='zeta=0.5')
plt.plot(ts,np.real(nfun14(ts)),label='zeta=1.2')
plt.legend()
```

oscylator harmoniczny z wymuszeniem

```python
x = sp.Function('x',real=True)
t = sp.Symbol('t',real=True)
w0 = sp.Symbol('omega_0',real=True,positive=True)
w = sp.Symbol('omega',real=True)
m = sp.Symbol('m',real=True,positive=True)
F0 = sp.Symbol('F_0',real=True)


x0 = sp.Symbol('x_0',real=True)
v0 = sp.Symbol('v_0',real=True)
```

```python
sol2 = sp.dsolve(x(t).diff(t,2)+w0**2*x(t)-F0/m*sp.cos(w*t),0,ics={x(0):x0,x(t).diff(t).subs(t,0):v0})
sol2
```

```python
nfun2 = sp.lambdify(t,sol2.rhs.subs({x0:10,v0:-1,w0:2,w:10,m:1,F0:10}))
nfun22 = sp.lambdify(t,sol2.rhs.subs({x0:10,v0:-1,w0:2,w:2.2,m:1,F0:10}))
nfun23 = sp.lambdify(t,sol2.rhs.subs({x0:10,v0:-1,w0:2,w:0.1,m:1,F0:10}))
nfun24 = sp.lambdify(t,sol2.rhs.subs({x0:10,v0:-1,w0:2,w:1.9999999999999,m:1,F0:10}))
```

```python
ts = np.linspace(0,15,1000)
plt.plot(ts,np.real(nfun2(ts)),label='omega = 10')
plt.plot(ts,np.real(nfun22(ts)),label='omega = 2.2')
plt.plot(ts,np.real(nfun23(ts)),label='omega = 0.1')
plt.plot(ts,np.real(nfun24(ts)),label='omega = 1.9999999999999')
plt.legend()
```

tłumienie i wymuszenie

```python
x = sp.Function('x',real=True)
t = sp.Symbol('t',real=True)
w0 = sp.Symbol('omega_0',real=True,positive=True)
w = sp.Symbol('omega',real=True)
m = sp.Symbol('m',real=True,positive=True)
F0 = sp.Symbol('F_0',real=True)
z = sp.Symbol('zeta',real=True)

x0 = sp.Symbol('x_0',real=True)
v0 = sp.Symbol('v_0',real=True)
```

```python
sol3 = sp.dsolve(x(t).diff(t,2)+2*z*w0*x(t).diff(t)+w0**2*x(t)-F0/m*sp.cos(w*t),0,ics={x(0):x0,x(t).diff(t).subs(t,0):v0})
sol3
```

```python
nfun3 = sp.lambdify(t,sol3.rhs.subs({x0:1,v0:-1,w0:2,w:1,m:1,F0:2,z:0.04}))
nfun32 = sp.lambdify(t,sol3.rhs.subs({x0:1,v0:-1,w0:2,w:1.99,m:1,F0:2,z:0.1}))
nfun33 = sp.lambdify(t,sol3.rhs.subs({x0:1,v0:-1,w0:2,w:1,m:1,F0:2,z:1.5}))
nfun34 = sp.lambdify(t,sol3.rhs.subs({x0:1,v0:-1,w0:2,w:1,m:1.99,F0:2,z:10}))
```

```python
ts = np.linspace(0,15,1000)
plt.plot(ts,np.real(nfun3(ts)),label='w=1 z=0.04')
plt.plot(ts,np.real(nfun32(ts)),label='w1.99= z=0.1')
plt.plot(ts,np.real(nfun33(ts)),label='w=1 z=1.5')
plt.plot(ts,np.real(nfun34(ts)),label='w=1.99 z=10')
plt.legend()
```
